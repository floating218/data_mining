```python
import pandas as pd

df=pd.read_csv('maryland_accident_dataset_by_report_no.csv').set_index('REPORT_NO')
df
```

```python
mycolumns=[
'COLLISION_TYPE_CODE',
'AREA_DAMAGED_CODE_MAIN',
'MOVEMENT_CODE'
]
```

```python

result_df=pd.read_excel('6_result_df.xlsx')

firstgroup=list(result_df.kmeans_labels)
firstgroup

```

```python
#df 



def return_df(df,groupname):
    global firstgroup
    df['firstgroup']=firstgroup
    df=df[df['firstgroup']==int(groupname)]

    # id  , person type D  
    df = df.drop(['PERSON_TYPE', 'PERSON_ID','VEHICLE_ID'], axis=1)

    #   .
    #df=df.drop(['RD_COND_CODE','SAF_EQUIP_CODE','EQUIP_PROB_CODE'],axis=1)

    #area_direction .
    area_direction=list(df.AREA_DAMAGED_CODE_MAIN).copy()
    for i in range(len(area_direction)):
        if area_direction[i][:-8] in ['Eleven','Twelve','One']:
            area_direction[i]='front'
        elif area_direction[i][:-8] in ['Two','Three','Four']:
            area_direction[i]='right'
        elif area_direction[i][:-8] in ['Five','Six','Seven']:
            area_direction[i]='back'
        elif area_direction[i][:-8] in ['Eight','Nine','Ten']:
            area_direction[i]='left'
        else:
            area_direction[i]='wrong'
    df['AREA_DAMAGED_CODE_MAIN']=area_direction

    #injury    
    #y  :fatal 1, no injury 0,  0.5 
    injury_str=list(df['INJ_SEVER_CODE'])
    for i in range(len(injury_str)):
        if injury_str[i]=='Fatal Injury':
            injury_str[i]=30.0
        elif injury_str[i]!='No Injury':
            injury_str[i]=15
        else:
            injury_str[i]=0
    df['INJURY']=injury_str 

    #month  .
    months=list(map(lambda x:x[5]+x[6],df.TIME))
    for i in range(len(months)):
        if months[i][0]=='0':
            months[i]=int(months[i][1])
        else:
            months[i]=int(months[i])
    for i in range(len(months)):
        if months[i]<4:
            months[i]=0
        elif months[i]<7:
            months[i]=1
        elif months[i]<10:
            months[i]=2
        else:
            months[i]=3
    df['MONTH']=months

    ages=list(df.AGE)
    for i in range(len(ages)):
        if ages[i]<30:
            ages[i]=0
        elif ages[i]<40:
            ages[i]=1
        elif ages[i]<50:
            ages[i]=2
        elif ages[i]<60:
            ages[i]=3
        else:
            ages[i]=4
    df['AGE']=ages
    return df

def preprocess(df,col_list):
    
    #1.  
    p_df=df.copy()    
    num_columns=['TIME','AGE','INJURY','MONTH']
    p_df=p_df.drop(num_columns,axis=1)
    
    p_df=p_df[col_list]

    #2.  
    dummies=list(p_df.columns)
    dummied = pd.get_dummies(p_df, columns=dummies)
    try:
        dummied=dummied.drop(['CDL_FLAG_N'],axis=1)
    except:
        pass
    try:
        dummied=dummied.drop(['FAULT_FLAG_N'],axis=1)
    except:
        pass
    try:
        dummied=dummied.drop(['HIT_AND_RUN_FLAG_N'],axis=1)
    except:
        pass
    try:
        dummied=dummied.drop(['C_M_ZONE_FLAG_N'],axis=1)
    except:
        pass
    p_df=dummied
    #--      .

    #4.   
    p_df['MONTH']=df['MONTH']
    p_df['AGE']=df['AGE']
    p_df['INJURY']=df['INJURY']

    #5.   
    scalings=['AGE','MONTH','INJURY']
    p_df[scalings]=(p_df \
        [scalings]-p_df[scalings].min())\
        /(p_df[scalings].max()-\
          p_df[scalings].min())
        

    return p_df

from sklearn.cluster import KMeans
def k_means(p_df):
    n_cluster=3
    kmeans_model=KMeans(n_clusters=3)
    kmeans_model.fit(p_df)
    #kmeans_model.cluster_centers_ #113  , 113  * 
    result_df=p_df.copy()
    result_df['kmeans_labels2']=kmeans_model.labels_
    return result_df

def show_analysis(result_df):
    analysis=result_df.groupby('kmeans_labels2').mean()
    result_df.groupby('kmeans_labels2').count()
    return analysis.transpose()

from sklearn.model_selection import train_test_split
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz

def decisiontree(df,groupname,col_list):
    p_df=return_df(df,groupname)
    p_df=preprocess(p_df,col_list)
    result_df=k_means(p_df)
    X,y=result_df.drop('kmeans_labels2',axis=1),result_df.kmeans_labels2
    #train,test
    X_train,X_test,y_train,y_test=train_test_split\
        (X,y,test_size=0.25,random_state=77,stratify=y)
    dt_model=DecisionTreeClassifier(class_weight='balanced',min_samples_leaf=100,max_depth=6) 
    dt_model.fit(X=X,y=y)
    dt_model.predict(X=X_test)#
    dt_model.score(X=X_test,y=y_test) #.   
    cluster_names=['0','1','2','3','4']
    with open('test_tree.dot','w') as f:
        export_graphviz(dt_model,f,feature_names=X.columns,class_names=cluster_names)
    with open('test_tree.dot','r') as f:
        dot_graph=f.read()
    dot_graph
    graphviz.Source(dot_graph)

def datamining(df,groupname,col_list):
    p_df=return_df(df,groupname)
    orin=p_df.firstgroup
    p_df=preprocess(p_df,col_list)
    result_df=k_means(p_df)
    return (p_df,result_df,show_analysis(result_df),list(orin))

```
